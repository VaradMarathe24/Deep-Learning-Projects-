```python
import pandas as pd
data=pd.read_csv("breast_cancer.csv")
data.head()
```

```python
data = data.drop(["Unnamed: 32","id"],axis=1)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
data["diagnosis"]=enc.fit_transform(data["diagnosis"])
data.head()
```

```python
y=data["diagnosis"].values
x = data.iloc[:,1:].values
```

```python
y
```

```python
print(y.shape)
print(x.shape)
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
pd.DataFrame(x).describe()
```

```python
#Split data into training and testing data
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.80, random_state=1)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
data["diagnosis"].value_counts()
```

```python
pd.DataFrame(ytrain).value_counts()
```

```python
x.shape[1]
```

```python
import keras as tf
```

```python
model=tf.models.Sequential()
model.add(tf.layers.Dense(16,input_dim=x.shape[1],activation="relu")) #creates 2 layers #16 number of neurons in hidden layer
model.add(tf.layers.Dense(32,activation="relu")) ##second hidden layer
model.add(tf.layers.Dense(1,activation="sigmoid"))

##Complie
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(xtrain,ytrain,epochs=500)
```

```python
ypred=model.predict(xtest)
ypred=ypred.round() ##Rounding off
ypred ##First output is zero so correct prediction
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(ytest,ypred)
cm
## 00 01 10 11 Confusion matrix elements
```

```python
ac=accuracy_score(ytest,ypred)
ac
```
